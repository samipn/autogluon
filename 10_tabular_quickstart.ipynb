## Classification: Breast Cancer (binary)

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
from autogluon.tabular import TabularPredictor
from sklearn.model_selection import train_test_split

X, y = load_breast_cancer(return_X_y=True, as_frame=True)
df = X.copy()
df['target'] = y

train_df, val_df = train_test_split(df, test_size=0.2, random_state=0, stratify=df['target'])
predictor = TabularPredictor(label='target', eval_metric='roc_auc', path='ag_quick_cls/').fit(train_df, time_limit=300)
predictor.leaderboard(val_df, silent=True)
```

## Regression: California Housing

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
from autogluon.tabular import TabularPredictor
from sklearn.model_selection import train_test_split

cal = fetch_california_housing(as_frame=True)
df = cal.frame.copy()
df.rename(columns={'MedHouseVal': 'target'}, inplace=True)

train_df, val_df = train_test_split(df, test_size=0.2, random_state=0)
predictor = TabularPredictor(label='target', problem_type='regression', eval_metric='rmse', path='ag_quick_reg/').fit(train_df, time_limit=300)
predictor.leaderboard(val_df, silent=True)
```
